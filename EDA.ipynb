```python
#Calculamos la distribuci√≥n porcentual de los juegos por a√±o(Esto ayuda a ver si hay a√±os con muy pocos datos)

games_percentage = (games_per_year / games_per_year.sum()) * 100
print(games_percentage)
```

```python
#Cu√°nto tardan generalmente las nuevas plataformas en aparecer y las antiguas en desaparecer.
#Se calcula el tiempo entre plataformas (Frecuencia de aparici√≥n de nuevas plataformas)
#Ordenamos por plataforma y a√±o de lanzamiento
platform_years = df.groupby('platform')['year_of_release'].min().reset_index()

#Diferencia de a√±os entre generaciones
platform_years = platform_years.sort_values(by='year_of_release')
platform_years['years_since_last'] = platform_years['year_of_release'].diff()

#Promedio de tiempo entre lanzamientos
avg_years_between_platforms = platform_years['years_since_last'].mean()
print(f"En promedio, una nueva plataforma aparece cada {avg_years_between_platforms:.1f} a√±os.")
```

```python
#C√°lculo de tiempo de desaparici√≥n
#Encontrar el primer y √∫ltimo a√±o con ventas por plataforma
platform_sales_range = df.groupby('platform')['year_of_release'].agg(['min', 'max']).reset_index()

#Calcular el tiempo activo de cada plataforma
platform_sales_range['lifespan'] = platform_sales_range['max'] - platform_sales_range['min']

#Calcular el promedio de duraci√≥n de una plataforma en el mercado
avg_platform_lifespan = platform_sales_range['lifespan'].mean()
print(f"En promedio, una plataforma permanece activa por {avg_platform_lifespan:.1f} a√±os antes de desaparecer.")
```

Para responder esta pregunta de forma precisa, podemos aplicar un an√°lisis estad√≠stico que nos permita evaluar si las diferencias observadas en las ventas entre plataformas son estad√≠sticamente significativas o si podr√≠an deberse al azar.

Adem√°s, la propuesta de un an√°lisis estad√≠stico para determinar la significancia de las diferencias en ventas es un excelente siguiente paso para fundamentar mejor las conclusiones. 

<div class="alert alert-block alert-warning">
<b>Comentario del revisor (1ra Iteraci√≥n)</b> <a class=‚ÄútocSkip‚Äù></a>

Como sugerencia, cuando tengas gr√°ficas *box-plot* con cajas muy "estrechas" debido a la presencia de numerosos datos at√≠picos, considera incluir una versi√≥n adicional usando escala logar√≠tmica. Esto facilitar√° la comparaci√≥n y el an√°lisis de los datos.

```python
from scipy.stats import mannwhitneyu
from itertools import combinations

#Combinaciones de plataformas para comparaci√≥n de pares
platform_pairs = list(combinations(top_platforms, 2))

for pair in platform_pairs:
    sales_1 = top_platforms_df[top_platforms_df['platform'] == pair[0]]['total_sales']
    sales_2 = top_platforms_df[top_platforms_df['platform'] == pair[1]]['total_sales']
    
    # Prueba de Mann-Whitney U
    stat, p_value = mannwhitneyu(sales_1, sales_2, alternative='two-sided')
    
    print(f'Comparaci√≥n: {pair[0]} vs {pair[1]} - Estad√≠stico: {stat}, p-valor: {p_value}')
    if p_value < 0.05:
        print("  -> Diferencia significativa.\n")
    else:
        print("  -> No hay diferencia significativa.\n")
```

<h1>Resultados</h1>
p-valor < 0.05:
Las plataformas en esa comparaci√≥n tienen diferencias significativas en las ventas.
p-valor ‚â• 0.05:
Las ventas no difieren significativamente entre esas dos plataformas.

```python
#Para comparar las ventas de los mismos juegos en diferentes plataformas (PS4, PC, y Xbox One)
#Analizamos c√≥mo las ventas globales var√≠an entre estas plataformas para los juegos que est√°n disponibles en varias de ellas.

#Filtrar los juegos que est√°n disponibles en al menos dos plataformas
platforms_of_interest = ['PS4', 'PC', 'Xbox One']
multiplatform_games = df[df['platform'].isin(platforms_of_interest)]

#Agrupar por nombre de juego y plataforma para obtener las ventas por plataforma
sales_by_platform = multiplatform_games.groupby(['name', 'platform'])[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()

#Calculamos las ventas globales para cada juego en cada plataforma
sales_by_platform['total_sales'] = sales_by_platform[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)

#Comparamos las ventas entre plataformas para cada juego
#Filtra lmosos juegos que est√°n en al menos 2 plataformas
multiplatform_sales_comparison = sales_by_platform.reset_index()
multiplatform_sales_comparison = multiplatform_sales_comparison[multiplatform_sales_comparison.groupby('name')['platform'].transform('nunique') >= 2]

#Resultado de la comparaci√≥n de ventas
print(multiplatform_sales_comparison[['name', 'platform', 'total_sales']])


```

<h1>PRUEBAS DE HIP√ìTESIS</h1>


```python
#Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

from scipy.stats import mannwhitneyu
import pandas as pd

#Las calificaciones de usuarios para Xbox One y PC
xbox_scores = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_scores = df[df['platform'] == 'PC']['user_score'].dropna()

#Las calificaciones a valores num√©ricos
xbox_scores = pd.to_numeric(xbox_scores, errors='coerce').dropna()
pc_scores = pd.to_numeric(pc_scores, errors='coerce').dropna()

#Prueba de Mann-Whitney U
stat, p_value = mannwhitneyu(xbox_scores, pc_scores, alternative='two-sided')

#Resultados
print(f'Estad√≠stico de prueba: {stat}')
print(f'p-valor: {p_value}')

#Interpretaci√≥n
if p_value < 0.05:
    print("Existe una diferencia significativa en las calificaciones promedio de los usuarios entre Xbox One y PC.")
else:
    print("No hay una diferencia significativa en las calificaciones promedio de los usuarios entre Xbox One y PC.")
```

<h1>An√°lisis del Resultado de la Prueba de Hip√≥tesis</h1>
Estad√≠stico de prueba: 51080.5
p-valor: 1.24e-08 (equivalente a 0.0000000124)

#Resultado
Rechazo de la Hip√≥tesis Nula (H‚ÇÄ):
El p-valor obtenido (1.24e-08) es mucho menor que el umbral com√∫nmente utilizado de 0.05.
Esto significa que rechazamos la hip√≥tesis nula y aceptamos la hip√≥tesis alternativa.
Por lo tanto, hay evidencia estad√≠sticamente significativa que indica que las calificaciones promedio de los usuarios en Xbox One y PC son diferentes.    

Es posible que los usuarios de una plataforma sean m√°s cr√≠ticos o indulgentes que los de la otra. Por ejemplo:
Si las calificaciones de PC son m√°s bajas, podr√≠a deberse a problemas de optimizaci√≥n o diferencias en la experiencia de juego.
Si las calificaciones de Xbox One son m√°s bajas, podr√≠a estar relacionado con expectativas de los usuarios.

Juegos mal optimizados en una plataforma espec√≠fica podr√≠an generar peores rese√±as. Los jugadores de PC, por ejemplo, suelen ser m√°s exigentes con el rendimiento y la personalizaci√≥n.

```python
#Para entender qu√© plataforma tiene mejores calificaciones, podemos calcular la media o mediana de las calificaciones de usuarios:
print("Media de calificaciones en Xbox One:", xbox_scores.mean())
print("Media de calificaciones en PC:", pc_scores.mean())

print("Mediana de calificaciones en Xbox One:", xbox_scores.median())
print("Mediana de calificaciones en PC:", pc_scores.median())
```

```python
#Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes.
import pandas as pd
from scipy.stats import mannwhitneyu

# Filtrar los datos para los g√©neros Acci√≥n y Deportes
action_scores = df[df['genre'] == 'Action']['user_score']
sports_scores = df[df['genre'] == 'Sports']['user_score']

# Convertir las calificaciones a valores num√©ricos (algunas pueden ser 'tbd' o NaN)
action_scores = pd.to_numeric(action_scores, errors='coerce').dropna()
sports_scores = pd.to_numeric(sports_scores, errors='coerce').dropna()

stat, p_value = mannwhitneyu(action_scores, sports_scores, alternative='two-sided')

print(f'Estad√≠stico de prueba: {stat}')
print(f'p-valor: {p_value}')

```
